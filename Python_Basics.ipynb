ii) **Five types of variables**
- <font color = 'blue'> Number 
  - arithmatical values
    - Integer (e.g., -3, -2, -1, 0, 1, 2, 3, etc.)
    - float (e.g., -3.0, -2.2 -1.0, 1.527, 3.14, etc.)

- <font color = 'blue'> String
  - any strings except for numbers 
    - e.g., 't', 'apple', 'extra', '12345', 'spelling'

- <font color = 'blue'> List
  - Items are listed in a pair of squared brackets.
  - Items are separated by commas. 
  - e.g., [-25, 0, 230, 3.14, 'an', 'earth', 'Mars', 'Jupiter]

- Tuple
  - Immutable connected items in a pair of brackets.
  - It is especially useful to keep the original, not making unwanted changes by mistake.
  - Items are separated by commas.  

- <font color = 'blue'> Dictionary
  - Key-value paris are related using a colon in a pair of curley brackets. 
  - e.g., {'apple': 5, 'pear':4, 'watermelon':3, 'lemon':2, 'cherry': 1}
